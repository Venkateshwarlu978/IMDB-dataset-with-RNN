```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

model = tf.keras.models.load_model("sentiment_bilstm_imdb.keras")

MAX_FEATURES = 10000
MAX_LEN = 500

word_index = imdb.get_word_index()

def encode_review(text):
    words = text.lower().split()
    encoded = [word_index.get(w, 2) for w in words if word_index.get(w, 2) < MAX_FEATURES]
    return pad_sequences([encoded], maxlen=MAX_LEN)

reviews = [
    "This movie was amazing and fantastic",
    "Worst movie ever boring and terrible"
]

for r in reviews:
    score = model.predict(encode_review(r))[0][0]
    print(r, "->", round(score, 3))
```
